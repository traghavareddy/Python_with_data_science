Regular expressions are used in Natural language processing to search or find a pattern in a text. 
A regular expression is a sequence of characters that defines a search pattern, useful for matching, searching, validating, and replacing text

```python
import re
```

```python
text = "I am doing well and I want to go home to study and sleep"
```

```python
# objective is to find all "to" in the text
re.findall("to",text)
```

```python
re.findall("study",text)
```

```python
re.findall("poem",text)
```

### Regular Expressions are used for finding patterns in a text

#### Example: Find phone numbers, fine email address

```python
text = "I like to buy Apple products while travelling, their quality is good like the actual apple fruit and apple is tempting"
```

```python
re.findall(r'[Aa]pple',text)
```

```python
re.findall(r'[A-Za-z]*ing',text)
```

```python
text2 = "I got two apples at Rs 90 and two bananas at Rs 20 alongwith an actual apple watch worth 50000 Rs"
```

```python
print(re.findall(r'\d{1,5}',text2))
```
