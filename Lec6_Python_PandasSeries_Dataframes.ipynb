## Merge Function

```python
df5 = pd.DataFrame({"ID":[11,22,33,44],"Name":["Alice","Bob","Charlie","David"],"Salary":[25000,27000,26000,50000]})
df5
```

```python
df6 = pd.DataFrame({"ID":[33,44,55,66],"Name":["Charlie","David","Eve","Frank"],"Age":[25,27,26,29]})
df6
```

```python
innerj = pd.merge(df5,df6,on="ID",how="inner")
innerj
```

```python
innerj_clean  = innerj.drop(["Name_y"],axis=1)
innerj_clean
```

```python
innerj_clean.columns = ["ID","Name","Sal","Age"]
innerj_clean
```

```python
leftj = pd.merge(df5,df6,on="ID",how="left")
leftj
```

```python
rightj = pd.merge(df5,df6,on="ID",how="right")
rightj
```

```python
outerj = pd.merge(df5,df6,on="ID",how="outer")
outerj
```

# Groupby Function

```python
# Customer DataFrame
data_cust = pd.DataFrame({
    "cust_id": [1, 2, 3, 4],
    "cust_name": ["Alka", "Ravi", "Neha", "Simran"],
    "region": ["North", "South", "North", "East"]
})

# Order DataFrame
data_order = pd.DataFrame({
    "order_id": [101, 102, 103, 104, 105],
    "cust_id": [1, 2, 1, 3, 4],
    "amount": [250, 450, 300, 200, 500]
})

data_cust

```

```python
data_order
```

```python
#Example 1: Group orders by cust_id and sum amounts

order_summary = data_order.groupby("cust_id")["amount"].sum()
order_summary
```

```python
#Example 2: Group customers by region and count them - How many customers per region

cust_region = data_cust.groupby("region")["cust_id"].count()
cust_region
```

```python
merged = pd.merge(data_cust, data_order, on="cust_id", how="inner")
merged
```

```python
# Groupby with aggregate function

cust_summary = merged.groupby("cust_id").agg({"amount": ["sum", "mean", "count", "max", "min"],"order_id":["count"]})
cust_summary
```

```python
merged.groupby("region").groups
```
