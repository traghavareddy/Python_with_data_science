# Conditional Statements: if , elif, else

```python
a = 9
b = 10
c = 5

print(a)
print(b)
print(c)
```

```python
if a>b:
    print(a,"is less than",b)
    print("End of If block")
else:
    print("Either b is less than or b is equal to a")
```

```python
a = 99
b = 10
c = 55
if (a<b) & (a<c):
    print(a,"is the smallest")
    print("this is the first block")
    f = a*9*7 #any operation or print can be written in a code block
    print(f)
elif (b<a) & (b<c):
    print(b,"is the smallest")
    print("code block")
else:
    print(c,"is the smallest")
```

```python
# lambda single line expressions for simple tasks - a substitute for functions
```

```python
square = lambda x:x*x
```

```python
square(7)
```

```python
def square1(x):
    return x*x #return function brings the control back to where the function is called
```

```python
c = square1(7)
print(c)
```

```python
s = lambda x,y:x*y
s(5,6)
```

```python
# map function: apply a certain function to all elements of a sequence: syntax is map(function,sequence)


num1 = [6,1,4,5,9]
list(map(square,num1))
```

```python
num1 = [6,1,4,5,9]
list(map(lambda x:x*x,num1))
```

# Loop control statements

```python
# break, continue, pass
```

```python
list(range(10))
```

```python
# break exits the loop:

for n in range(10):
    if n == 5:
        break
    else:
        print("**",n,"**")
```

```python
# continue skips to next iteration:

for n in range(10):
    if n == 5:
        continue
    else:
        print("**",n,"**")
```

```python
# pass is used as a placeholder when syntax needs a statement but you don
x = 10
if x>0:
    pass
else:
    print("Negative or Zero")
```
