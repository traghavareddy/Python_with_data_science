## Example 1: Simple class with one function

```python
# Class definition using class keywork
class employee:
    def work(self): #self is used by python to refer to the current object
        print("Employee works like me!")
```

```python
# object of the class

e1 = employee()
```

```python
# function defined within the class can be access only via object of the class

e1.work()
```

## Example 2: Class with a constructor used to assign values to an object

```python
#class definition
class employee2():
    def __init__(self,name,age,id,salary): #Contructor is defined by keyword __init__ and executes when you create an object of the class
        self.name = name
        self.age = age
        self.salary = salary
        self.id = id
        print("Employee Created Successfully!")
```

```python
# object creation

na = input("Enter the name:")
ag = input("Enter the age")

emp1 = employee2(na,ag,123,40000) #emp1 is the name of the object getting created
```

```python
emp1.name
```

```python
emp1.age
```

```python
emp1.salary
```

```python
emp2 = employee2("Abc",27,1234,50000)
```

```python
emp2.name
```

```python
emp1.name
```

## Example 3: class with constructor and a function

```python
class education:
    def __init__(self,nameofstudent,subj):
        self.name = nameofstudent
        self.subj = subj
        print("Hello, and welcome to the class",self.name)
    def subject(self):
        print("My favorite subject is:",self.subj)
```

```python
s1 = education("Alka","Maths")
```

```python
s1.subject()
```

## Inheritance

```python
class Animal(): #parent class
    def speak(self):
        print("An animal is speaking....")

class Dog(Animal): #chile class
    def bark(self):
        print("Woof Woof! Dog Barks...")
```

```python
a1 = Animal()
```

```python
a1.speak()
```

```python
d = Dog()
```

```python
d.bark()
```

```python
d.speak()
```

## Multiple Inheritance:


```python
class Father():
    def skills(self):
        print("Gardening...")

class Mother():
    def hobbies(self):
        print("Painting...")

class Child(Father,Mother):
    def talent(self):
        print("Singing...")
```

```python
elder = Child()
```

```python
elder.talent()
```

```python
elder.skills()
```

```python
elder.hobbies()
```

## Polymorphism - Ability of an object to be processed in many forms

```python
class Father():
    def skills(self):
        print("Gardening...")

class Mother():
    def skills(self):
        print("Painting...")

class Child(Father,Mother):
    def skills(self):
        print("Singing...")
```

```python
younger = Child()
```

```python
younger.skills() #Method overriding - child's method will be overrideen if it has the same name as parent
```

```python
m = Mother()
m.skills()
```

## Multilevel:

```python
class Grandfather():
    def skills1(self):
        print("Gardening...")

class Father(Grandfather):
    def skills2(self):
        print("Painting...")

class Child(Father):
    def skills3(self):
        print("Singing...")
```

```python
cc = Child()
cc.skills3()
```

```python
cc.skills2()
```

```python
cc.skills1()
```
