### Variable: Variable is a named storage in memory used to hold data

#### Python is a dynamically typed language i.e no need to explicitly declare the data type of the variable, it automatically understands when values are assigned to it during execution

```python
# Create a variable "name" with the value "Raj" assigned

name = "Raj" #Double quotes indicates python that the variable is a string type
print("Hello",name)
```

```python
# Create an integer variable "num" with the value 800 assigned to it

num = 800 #Since there are no decimal points and no double brackets, python treats it like integer data type
print(num)
```

```python
# Create a variable "Salary" with the value 3000.50 assigned to it

Salary = 3000.50 #Decimal value indicates python that Salary is a float type
print(Salary)
```

## Mutable and Immutable Data Types in Python

###### Mutable data types are the data types whose values can be changed after they are created - Lists, Dictionaries, Sets
###### Immutable data types are the ones whose values cannot be changed once created - Numbers, Strings, Tuples

```python
# Numerical Data Type

x = 9
y = 7.8
z = 7+6j

print(type(x))
print(type(y))
print(type(z))
```

```python
#check the instance:

print(isinstance(x,int))
print(isinstance(x,float))
```

```python
# Type conversion: IMPORTANT to understand that convert logically

a = "8"
b = 9.5
c= 6
print(int(a))
print(int(b))
print(float(c))
```

```python
# String

d = "Alka Pandey"
print(d[2])

#Can we change anything in the string?

d[2] = "i" #Error because strings are immutable data types
```

```python
# Tuples: Tuple is an ordered collection of items enclosed within round brackets - immutable data type

t1 = (11,28,39)
print(t1)
id(t1)
```

#### Mutable Data Types:

```python
# Lists are ordred collection of elements enclosed within square brackets

l1 = [11,2,3,5,7,8.7,"N"]
print(l1)

#Access elements using indexing

l1[0]
```

```python
# Can you change elements in list?

l1[2] = 900
print(l1)
```

```python
# Dictionary: collection key-value pairs enclosed in curly brackets which are mutable

d = {"K1":"V1","Name":"AP","Age":30}
print(d)
```
